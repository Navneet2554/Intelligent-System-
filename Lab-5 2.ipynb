```python
from sklearn import preprocessing
```

```python
weather=["Sunny","Sunny","Overcast","Rainy","Rainy","Rainy","Overcast","Sunny","Sunny","Rainy","Sunny","Overcast","Overcast","Rainy"]

temp=["hot","hot","hot","mild","cool","cool","cool","mild","cool","mild","mild","mild","hot","mild"]

play=["No","No","Yes","yes","yes","No","yes","No","yes","yes","yes","yes","yes","No"]
```

```python
le=preprocessing.LabelEncoder()
```

```python
weather_encoder=le.fit_transform(weather)
print(weather_encoder)
```

```python
temp_encoder=le.fit_transform(temp)
print(temp_encoder)
```

```python
play_encoder=le.fit_transform(play)
print(play_encoder)
```

```python
label=le.fit_transform(play)

feature=list(zip(weather_encoder,temp_encoder))
print(feature)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model=KNeighborsClassifier(n_neighbors=3)
model.fit(feature,label)
```

```python
predicted=model.predict([[0,2]])
predicted
```

```python
from sklearn import datasets
```

```python
wine=datasets.load_wine()
print(wine.feature_names)
print(wine.target_names)
print(wine.data[0,5])
```

```python
print(wine.data.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(wine.data,wine.target,test_size=0.3)
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
print(y_pred)
```

```python
from sklearn import metrics
```

```python
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```

```python
import numpy as np
df.isna().sum()
```

```python
droped=df.dropna()
droped
```

```python
species_encoder=le.fit_transform(droped["species"])
species_encoder
```

```python
island_encoder=le.fit_transform(droped["island"])
s_encoder=le.fit_transform(droped["sex"])
```

```python
feature=list(zip(species_encoder,island_encoder))
x_train,x_test,y_train,y_test=train_test_split(feature,s_encoder,test_size=0.3)

knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
print("Accuracy :",metrics.accuracy_score(y_test,y_pred))
```
